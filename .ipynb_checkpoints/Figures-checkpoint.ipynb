# Simulating a Geometric Brownian Motion: Defaults at Maturity

```python
V0 = 100
rf = 0.035 # 3.5 %
rm = 0.105 # 10.5% 
beta = 0.8
sigma_i = 0.25
sigma_m = 0.14
ttm = 5
n = 10000
```

```python
mu = rf + beta * (rm - rf)
sigma = ott.sigma_beta_adj(beta, sigma_m, sigma_i)
default_prob_B = DefTable.loc["B", ttm] / 100
face_value = ott.facevalue_from_probability(default_prob_B, V0, ttm, mu, sigma)
market_value = ott.mv_bond(V0, face_value, ttm, rf, sigma)
```

```python
simulations = ott.Stochastic_Model(T = ttm, freq = 50, seed = 1234)
gbm_path = simulations.GBM(n=n, v_0=V0, mu=mu, sigma=sigma)

data = pd.DataFrame(gbm_path)
data.T.head(6)
```

```python
import scipy.stats as stats
```

```python
# Set up the axes with gridspec
fig = plt.figure(figsize=(textwidth, 7.5*cm), clear = True)
grid = plt.GridSpec(1, 2, wspace=0.008, width_ratios=[3,1])

palette = sns.color_palette("summer", as_cmap=True)

main_ax = fig.add_subplot(grid[0])

sec_ax = fig.add_subplot(grid[1], xticks=[], yticks=[], frameon=False) #sharey=main_ax)

# Get Min and Max
dmin = data.iloc[-1].min()
dmax = np.minimum(data.iloc[-1].max(),600)
default_barrier = face_value

# Adjust left plot
main_ax.plot(data.iloc[:,0:250], linestyle='-', linewidth=0.5)
main_ax.margins(x=0)
main_ax.set_ylim(dmin, dmax)
main_ax.set_yticks([0,100,200,300,400,500,600])
main_ax.set_yticklabels([0,100,200,300,400,500,600])
main_ax.axhline(y=default_barrier, c='k', linestyle='-.', linewidth=1.25)
main_ax.set_xticks([0,50,100,150,200,250], minor=False)
main_ax.set_xticklabels([0,1,2,3,4,5], minor=False)
main_ax.set_xlabel('Year')
main_ax.set_ylabel('Firm Value')

# Adjust right plot
sec_ax.set_ylim(0, dmax)
sec_ax.margins(x=0, y=0)
sec_ax.set_yticklabels([])
sec_ax.hist(data.iloc[-1],
            range=(dmin, dmax),
            bins = np.minimum(int(n/7),50),
            histtype='bar',
            orientation='horizontal',
            align = 'mid',
            density=True,
            stacked =True,
            color=palette(1))
sec_ax.axhline(y=default_barrier, c='k', linestyle='-.', linewidth=1.25, label="Face Value")

lin_ax = np.linspace(dmin, dmax, 100)

# Create right plots
params = stats.lognorm.fit(data.iloc[-1], floc=0) # shape (standard deviation), loc and scale (mean)
pdf = stats.lognorm.pdf(lin_ax, s = params[0], loc = params[1], scale = params[2])
sec_ax.plot(pdf, lin_ax, color='red', linewidth=0.75)

labels = ["Default Barrier"]
handles, _ = main_ax.get_legend_handles_labels()

# Slice list to remove first handle
plt.legend(handles = handles[1:], labels = labels, frameon=False)

plt.show()

#plt.savefig("pic.pdf", format='pdf', bbox_inches='tight')
```

# Yields in Merton's Model

```python
v = 1
t = 1
rf = 0
values = np.linspace(0.000001,1,100)
```

```python
def yield_call(fv, vol):
    mv = ott.mv_bond(v, fv, t, rf, vol)
    return ott.zero_yield(mv, fv, t)
```

```python
plt.figure(figsize=(textwidth, 6*cm))
palette = sns.color_palette("summer", n_colors=6, as_cmap=True)

plt.plot(values, yield_call(values, 0.2)*100, color=palette(1), linestyle ='-.')
plt.plot(values, yield_call(values, 0.35)*100, color=palette(1), linestyle ='--')
plt.plot(values, yield_call(values, 0.5)*100, color=palette(1))

plt.legend(("$\sigma={}\%$".format(20),"$\sigma={}\%$".format(35),"$\sigma={}\%$".format(50)),loc="upper center",ncol=3, frameon=False)

plt.ylabel("Spread, $s(0,1)$")
plt.ylim(0,25)
plt.yticks(np.linspace(0, 25, 6),["0%","5%","10%","15%","20%","25%"])

plt.xlim([0,1])
plt.xlabel("Leverage $B\,/\,V_0$")

plt.tight_layout()

#plt.savefig("pic.pdf", format='pdf', bbox_inches='tight')

plt.show()
```
